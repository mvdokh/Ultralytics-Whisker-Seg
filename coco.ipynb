```python
import os
from pathlib import Path

def remove_img_prefix(folder_path):
    """
    Remove 'img' prefix from all .png files in the specified folder.
    
    Args:
        folder_path (str): Path to the folder containing images
    """
    folder = Path(folder_path)
    
    if not folder.exists():
        print(f"Error: Folder {folder_path} does not exist!")
        return
    
    # Find all PNG files that start with 'img'
    img_files = list(folder.glob("img*.png"))
    
    if not img_files:
        print(f"No PNG files with 'img' prefix found in {folder_path}")
        return
    
    print(f"Found {len(img_files)} files with 'img' prefix:")
    
    renamed_count = 0
    for img_file in img_files:
        # Get the filename without the 'img' prefix
        old_name = img_file.name
        new_name = old_name[3:]  # Remove first 3 characters ('img')
        new_path = img_file.parent / new_name
        
        # Check if target file already exists
        if new_path.exists():
            print(f"Warning: {new_name} already exists, skipping {old_name}")
            continue
        
        try:
            # Rename the file
            img_file.rename(new_path)
            print(f"Renamed: {old_name} -> {new_name}")
            renamed_count += 1
        except Exception as e:
            print(f"Error renaming {old_name}: {e}")
    
    print(f"\nSuccessfully renamed {renamed_count} files")

# Set the path to your images folder
image_folder = "/COCO/datasets/whisker-seg/images/train"

# Check if the folder exists and run the script
if os.path.exists(image_folder):
    print(f"Processing folder: {image_folder}")
    remove_img_prefix(image_folder)
else:
    print(f"Folder not found: {image_folder}")
    print("Current working directory:", os.getcwd())
    
    # Try relative path from current location
    relative_path = "datasets/whisker-seg/images/train"
    if os.path.exists(relative_path):
        print(f"Found relative path: {relative_path}")
        remove_img_prefix(relative_path)
    else:
        print("Please check the folder path and try again")
```

```python
import os
from pathlib import Path
from csv_to_ultralytics_enhanced_utils import csv_to_ultralytics_enhanced

def batch_process_whisker_data(base_directory, **kwargs):
    """
    Process all subdirectories in Whisker_Training_Data with enhanced CSV conversion.
    
    Args:
        base_directory (str): Path to the Whisker_Training_Data directory
        **kwargs: Additional arguments to pass to csv_to_ultralytics_enhanced
    """
    
    base_path = Path(base_directory)
    
    if not base_path.exists():
        print(f"Error: Directory {base_directory} does not exist!")
        return
    
    # Get all subdirectories
    subdirs = [d for d in base_path.iterdir() if d.is_dir()]
    
    if not subdirs:
        print(f"No subdirectories found in {base_directory}")
        return
    
    print(f"Found {len(subdirs)} directories to process:")
    for subdir in subdirs:
        print(f"  - {subdir.name}")
    
    print(f"\nStarting batch processing with settings:")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")
    print()
    
    # Track statistics
    successful = []
    failed = []
    
    # Process each subdirectory
    for i, subdir in enumerate(subdirs, 1):
        print(f"[{i}/{len(subdirs)}] Processing {subdir.name}...")
        
        try:
            # Check if directory has the expected structure
            images_dir = subdir / "images"
            labels_dir = subdir / "labels"
            
            if not images_dir.exists():
                print(f"  ⚠ Warning: No 'images' folder found in {subdir.name}")
                continue
                
            if not labels_dir.exists():
                print(f"  ⚠ Warning: No 'labels' folder found in {subdir.name}")
                continue
            
            # Run the enhanced conversion
            output_path = csv_to_ultralytics_enhanced(str(subdir), **kwargs)
            successful.append(subdir.name)
            print(f"  ✓ Successfully processed {subdir.name}")
            print(f"    Output: {output_path}")
            
        except Exception as e:
            failed.append((subdir.name, str(e)))
            print(f"  ✗ Failed to process {subdir.name}: {e}")
        
        print()  # Add spacing between directories
    
    # Print final summary
    print("="*60)
    print("BATCH PROCESSING SUMMARY")
    print("="*60)
    print(f"Total directories: {len(subdirs)}")
    print(f"Successfully processed: {len(successful)}")
    print(f"Failed: {len(failed)}")
    
    if successful:
        print(f"\nSuccessful:")
        for name in successful:
            print(f"  ✓ {name}")
    
    if failed:
        print(f"\nFailed:")
        for name, error in failed:
            print(f"  ✗ {name}: {error}")

# BATCH PROCESSING - Process all folders in Whisker_Training_Data
base_dir = "Whisker_Training_Data"

# Option 1: Use default settings for all folders
print("Processing all folders with default settings...")
batch_process_whisker_data(
    base_dir,
    single_class=True,        # All categories -> class 0
    epsilon_factor=0.002,     # Good balance of detail/simplification  
    buffer_pixels=3           # 3-pixel buffer around whiskers
)

# Option 2: Custom settings (uncomment to use)
# print("Processing all folders with custom settings...")
# batch_process_whisker_data(
#     base_dir,
#     single_class=True,        # All categories -> class 0
#     epsilon_factor=0.001,     # More detailed (less simplification)
#     buffer_pixels=5           # Larger buffer
# )
```

```python
import os
import shutil

def reorganize_whisker_seg_labels(base_dir):
    """
    For each subdirectory in whisker-seg:
    1. Delete the 'labels' folder
    2. Rename 'ultralytics_labels_enhanced' to 'labels'
    
    Args:
        base_dir (str): Path to the whisker-seg directory
    """
    
    # Track statistics
    total_dirs_processed = 0
    total_labels_deleted = 0
    total_renamed = 0
    errors = []
    
    # Get all subdirectories
    subdirs = [d for d in os.listdir(base_dir) 
               if os.path.isdir(os.path.join(base_dir, d))]
    
    print(f"Found {len(subdirs)} subdirectories to process")
    
    for subdir in subdirs:
        subdir_path = os.path.join(base_dir, subdir)
        labels_path = os.path.join(subdir_path, 'labels')
        enhanced_labels_path = os.path.join(subdir_path, 'ultralytics_labels_enhanced')
        
        print(f"\nProcessing {subdir}...")
        
        try:
            # Check if both folders exist
            labels_exists = os.path.exists(labels_path)
            enhanced_exists = os.path.exists(enhanced_labels_path)
            
            if not enhanced_exists:
                print(f"  Warning: 'ultralytics_labels_enhanced' folder not found in {subdir}")
                continue
            
            # Delete the original labels folder if it exists
            if labels_exists:
                shutil.rmtree(labels_path)
                print(f"  ✓ Deleted 'labels' folder")
                total_labels_deleted += 1
            else:
                print(f"  Note: 'labels' folder doesn't exist in {subdir}")
            
            # Rename ultralytics_labels_enhanced to labels
            os.rename(enhanced_labels_path, labels_path)
            print(f"  ✓ Renamed 'ultralytics_labels_enhanced' to 'labels'")
            total_renamed += 1
            total_dirs_processed += 1
            
        except Exception as e:
            error_msg = f"Error processing {subdir}: {str(e)}"
            errors.append(error_msg)
            print(f"  ✗ {error_msg}")
    
    # Print summary
    print(f"\n=== SUMMARY ===")
    print(f"Directories successfully processed: {total_dirs_processed}")
    print(f"Labels folders deleted: {total_labels_deleted}")
    print(f"Enhanced labels folders renamed: {total_renamed}")
    print(f"Errors encountered: {len(errors)}")
    
    if errors:
        print("\nErrors:")
        for error in errors:
            print(f"  - {error}")

# Set the whisker-seg directory path
whisker_seg_dir = "/home/mvdokh/tracking/COCO-HPCC/datasets/whisker-seg"

# Run the function
reorganize_whisker_seg_labels(whisker_seg_dir)
```

```python
import os
import shutil
import random
from pathlib import Path

def split_images_train_val(base_dir, train_ratio=0.8):
    """
    For each subdirectory in whisker-seg:
    1. Create train and val folders inside the images folder
    2. Move 80% of images to train folder and 20% to val folder
    
    Args:
        base_dir (str): Path to the whisker-seg directory
        train_ratio (float): Ratio of images for training (default: 0.8)
    """
    
    # Track statistics
    total_dirs_processed = 0
    total_images_moved = 0
    errors = []
    
    # Get all subdirectories
    subdirs = [d for d in os.listdir(base_dir) 
               if os.path.isdir(os.path.join(base_dir, d))]
    
    print(f"Found {len(subdirs)} subdirectories to process")
    print(f"Split ratio: {train_ratio*100:.0f}% train, {(1-train_ratio)*100:.0f}% val\n")
    
    for subdir in subdirs:
        subdir_path = os.path.join(base_dir, subdir)
        images_path = os.path.join(subdir_path, 'images')
        
        print(f"Processing {subdir}...")
        
        try:
            # Check if images folder exists
            if not os.path.exists(images_path):
                print(f"  Warning: 'images' folder not found in {subdir}")
                continue
            
            # Create train and val folders
            train_path = os.path.join(images_path, 'train')
            val_path = os.path.join(images_path, 'val')
            
            os.makedirs(train_path, exist_ok=True)
            os.makedirs(val_path, exist_ok=True)
            print(f"  ✓ Created train and val folders")
            
            # Get all image files (excluding the train and val directories)
            all_files = [f for f in os.listdir(images_path) 
                        if os.path.isfile(os.path.join(images_path, f)) and 
                        f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff', '.tif'))]
            
            if not all_files:
                print(f"  Warning: No image files found in {subdir}/images")
                continue
            
            # Shuffle the files for random split
            random.shuffle(all_files)
            
            # Calculate split index
            train_count = int(len(all_files) * train_ratio)
            
            # Split files
            train_files = all_files[:train_count]
            val_files = all_files[train_count:]
            
            # Move files to train folder
            for file in train_files:
                src_path = os.path.join(images_path, file)
                dst_path = os.path.join(train_path, file)
                shutil.move(src_path, dst_path)
            
            # Move files to val folder
            for file in val_files:
                src_path = os.path.join(images_path, file)
                dst_path = os.path.join(val_path, file)
                shutil.move(src_path, dst_path)
            
            print(f"  ✓ Moved {len(train_files)} images to train folder")
            print(f"  ✓ Moved {len(val_files)} images to val folder")
            
            total_images_moved += len(all_files)
            total_dirs_processed += 1
            
        except Exception as e:
            error_msg = f"Error processing {subdir}: {str(e)}"
            errors.append(error_msg)
            print(f"  ✗ {error_msg}")
    
    # Print summary
    print(f"\n=== SUMMARY ===")
    print(f"Directories successfully processed: {total_dirs_processed}")
    print(f"Total images moved: {total_images_moved}")
    print(f"Errors encountered: {len(errors)}")
    
    if errors:
        print("\nErrors:")
        for error in errors:
            print(f"  - {error}")

# Set random seed for reproducible splits
random.seed(42)

# Set the whisker-seg directory path
whisker_seg_dir = "/home/mvdokh/tracking/COCO-HPCC/datasets/whisker-seg"

# Run the function
split_images_train_val(whisker_seg_dir)
```

```python
import os
import shutil
from pathlib import Path

def split_labels_train_val(base_dir):
    """
    For each subdirectory in whisker-seg:
    1. Create train and val folders inside the labels folder
    2. Move label files to match the split of their corresponding image files
    
    Args:
        base_dir (str): Path to the whisker-seg directory
    """
    
    # Track statistics
    total_dirs_processed = 0
    total_labels_moved = 0
    errors = []
    unmatched_labels = []
    
    # Get all subdirectories
    subdirs = [d for d in os.listdir(base_dir) 
               if os.path.isdir(os.path.join(base_dir, d))]
    
    print(f"Found {len(subdirs)} subdirectories to process\n")
    
    for subdir in subdirs:
        subdir_path = os.path.join(base_dir, subdir)
        images_path = os.path.join(subdir_path, 'images')
        labels_path = os.path.join(subdir_path, 'labels')
        
        print(f"Processing {subdir}...")
        
        try:
            # Check if both images and labels folders exist
            if not os.path.exists(images_path):
                print(f"  Warning: 'images' folder not found in {subdir}")
                continue
                
            if not os.path.exists(labels_path):
                print(f"  Warning: 'labels' folder not found in {subdir}")
                continue
            
            # Check if train/val folders exist in images
            images_train_path = os.path.join(images_path, 'train')
            images_val_path = os.path.join(images_path, 'val')
            
            if not os.path.exists(images_train_path) or not os.path.exists(images_val_path):
                print(f"  Warning: train/val folders not found in {subdir}/images")
                continue
            
            # Create train and val folders in labels
            labels_train_path = os.path.join(labels_path, 'train')
            labels_val_path = os.path.join(labels_path, 'val')
            
            os.makedirs(labels_train_path, exist_ok=True)
            os.makedirs(labels_val_path, exist_ok=True)
            print(f"  ✓ Created train and val folders in labels")
            
            # Get image files from train and val folders
            train_images = [f for f in os.listdir(images_train_path) 
                           if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff', '.tif'))]
            val_images = [f for f in os.listdir(images_val_path) 
                         if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff', '.tif'))]
            
            # Create sets of image basenames (without extensions) for quick lookup
            train_basenames = {os.path.splitext(img)[0] for img in train_images}
            val_basenames = {os.path.splitext(img)[0] for img in val_images}
            
            # Get all label files (excluding the train and val directories)
            all_label_files = [f for f in os.listdir(labels_path) 
                              if os.path.isfile(os.path.join(labels_path, f)) and 
                              f.lower().endswith('.txt')]
            
            if not all_label_files:
                print(f"  Warning: No label files found in {subdir}/labels")
                continue
            
            train_labels_moved = 0
            val_labels_moved = 0
            unmatched_in_subdir = 0
            
            # Process each label file
            for label_file in all_label_files:
                label_basename = os.path.splitext(label_file)[0]
                src_path = os.path.join(labels_path, label_file)
                
                if label_basename in train_basenames:
                    # Move to train folder
                    dst_path = os.path.join(labels_train_path, label_file)
                    shutil.move(src_path, dst_path)
                    train_labels_moved += 1
                elif label_basename in val_basenames:
                    # Move to val folder
                    dst_path = os.path.join(labels_val_path, label_file)
                    shutil.move(src_path, dst_path)
                    val_labels_moved += 1
                else:
                    # No corresponding image found
                    unmatched_labels.append(f"{subdir}/{label_file}")
                    unmatched_in_subdir += 1
            
            print(f"  ✓ Moved {train_labels_moved} labels to train folder")
            print(f"  ✓ Moved {val_labels_moved} labels to val folder")
            if unmatched_in_subdir > 0:
                print(f"  ⚠ {unmatched_in_subdir} labels had no corresponding images")
            
            total_labels_moved += (train_labels_moved + val_labels_moved)
            total_dirs_processed += 1
            
        except Exception as e:
            error_msg = f"Error processing {subdir}: {str(e)}"
            errors.append(error_msg)
            print(f"  ✗ {error_msg}")
    
    # Print summary
    print(f"\n=== SUMMARY ===")
    print(f"Directories successfully processed: {total_dirs_processed}")
    print(f"Total labels moved: {total_labels_moved}")
    print(f"Unmatched labels: {len(unmatched_labels)}")
    print(f"Errors encountered: {len(errors)}")
    
    if errors:
        print("\nErrors:")
        for error in errors:
            print(f"  - {error}")
    
    if unmatched_labels:
        print(f"\nUnmatched labels (first 10):")
        for label in unmatched_labels[:10]:
            print(f"  - {label}")
        if len(unmatched_labels) > 10:
            print(f"  ... and {len(unmatched_labels) - 10} more")

# Set the whisker-seg directory path
whisker_seg_dir = "/home/mvdokh/tracking/COCO-HPCC/datasets/whisker-seg"

# Run the function
split_labels_train_val(whisker_seg_dir)
```
